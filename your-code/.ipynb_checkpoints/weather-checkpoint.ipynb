## Global Historical Climatology Network Dataset
Variables are stored in both rows and columns
This dataset represents the daily weather records for a weather station (MX17004) in Mexico for five months in 2010.

```python
import pandas as pd
import datetime as dt
weather = pd.read_csv('weather-raw.csv')
```

```python
w = weather.melt(id_vars=['id','year','month','element'], var_name='day',value_name='temp')
w.head()
```

```python
w2 = w.pivot_table(index=['id','year','month','day'], values='temp', columns='element').reset_index().assign(day= w.day.apply(lambda x: x.replace('d','')).astype('int'))
```

```python
w2['date'] = pd.to_datetime(w2.apply(lambda row: dt.date(row['year'], row['month'], row['day']), axis=1))
```

```python
w2.set_index('date', inplace=True)
```
